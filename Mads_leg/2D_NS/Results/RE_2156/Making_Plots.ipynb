```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Create the aggregated loss plot

```python
# Load the data
data_std = pd.read_csv("/home/madshh7/PINN_Bachelor/Mads_leg/2D_NS/Results/RE_2156/Loss/Pressure_val.csv")
data_SAQR = pd.read_csv("/home/madshh7/PINN_Bachelor/Mads_leg/2D_NS/Results/RE_2156/Loss/SAQR_RE_2553.csv") 
data_SoftAdapt = pd.read_csv("/home/madshh7/PINN_Bachelor/Mads_leg/2D_NS/Results/RE_2156/Loss/SA_RE_2553.csv")
```

```python
# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))
ax.semilogy(data_SAQR["Step"], data_SAQR["Value"], label="SoftAdapt + Quasi-random", alpha=0.5)
ax.semilogy(data_SoftAdapt["Step"], data_SoftAdapt["Value"], label="SoftAdapt", alpha=0.5)
ax.semilogy(data_std["Step"], data_std["Value"], label="Standard")
# Add labels and title
ax.set_xlabel("Step", fontsize=12)
ax.set_ylabel("Value", fontsize=12)
ax.set_title("Aggregated loss, RE$=2553$", fontsize=14)

# Add grid and legend
ax.grid(True, which='both', linestyle='--', linewidth=0.5)
ax.legend()
plt.show()
```

### Plot the relative error for $p$

```python
# Load Relative Errors
RE_std = pd.read_csv("/home/madshh7/PINN_Bachelor/Mads_leg/2D_NS/Results/RE_2156/Pressure/STD_RE_2553.csv")
RE_SA = pd.read_csv("/home/madshh7/PINN_Bachelor/Mads_leg/2D_NS/Results/RE_2156/Pressure/SA_RE_2553 (1).csv")
RE_SAQR = pd.read_csv("/home/madshh7/PINN_Bachelor/Mads_leg/2D_NS/Results/RE_2156/Pressure/SAQR_RE_2553 (1).csv")
```

```python
# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))
ax.semilogy(RE_SAQR["Step"], RE_SAQR["Value"], label="SoftAdapt + Quasi-random")
ax.semilogy(RE_SA["Step"], RE_SA["Value"], label="SoftAdapt")
ax.semilogy(RE_std["Step"], RE_std["Value"], label="Standard")

# Add labels and title
ax.set_xlabel("Iteration", fontsize=12)
ax.set_ylabel("Relative error", fontsize=12)
ax.set_title(f"Relative pressure error, RE$=2553$", fontsize=14)

# Add grid and legend
ax.grid(True, which='both', linestyle='--', linewidth=0.5)
ax.legend()
plt.show()
```
